```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
#importing chromedriver
driver = webdriver.Chrome(r"C:\Users\ashis\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

```python
location = []
Company = []
Model = []
ModelType = []
Year = []
Kms      = []
Fuel     = []
Car_Type = []
Color = []
No_of_owner = []
RTO      = []
Transmission = []
Insurance = []
Price = []
```

```python
def process(url):    
    driver.implicitly_wait(2)
    try:
        SCROLL_PAUSE_TIME = 0.5
        driver.get(url)
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        urls = []
        url_link=driver.find_elements_by_xpath("//table[@class='grid-12 omega alpha card-detail-block__data-table']/tbody/tr/td/h2/a")#collecting urls
        for i in url_link:
            urls.append(i.get_attribute('href'))
    finally:
        print("Done")
        
    for i in urls:
        driver.get(i)
        time.sleep(2)
        loc=driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[3]/a/span") #item name
        location.append(loc.text)
        com=driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[4]/a/span") #item name
        Company.append(com.text)
        mod=driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[5]/a/span") #item name
        Model.append(mod.text)
        modtype=driver.find_element_by_xpath("//ul[@class='special-skin-text']/li[6]/span[2]") #item name
        ModelType.append(modtype.text)
        yer=driver.find_element_by_xpath("//ul[@class='basic-details']/li[1]/span[2]") #item name
        Year.append(yer.text)
        km=driver.find_element_by_xpath("//ul[@class='basic-details']/li[2]/span[2]") #item name
        Kms.append(km.text)
        ful=driver.find_element_by_xpath("//ul[@class='basic-details']/li[3]/span[2]") #item name
        Fuel.append(ful.text)
        crtype=driver.find_element_by_xpath("//ul[@class='basic-details']/li[4]/span[2]") #item name
        Car_Type.append(crtype.text)
        trans=driver.find_element_by_xpath("//div[@class='cw-tabs-data']/div/ul/li[5]/div[2]") #item name
        Transmission.append(trans.text)
        col=driver.find_element_by_xpath("//div[@class='cw-tabs-data']/div/ul/li[7]/div[2]") #item name
        Color.append(col.text)
        owner=driver.find_element_by_xpath("//div[@class='cw-tabs-data']/div/ul/li[9]/div[2]") #item name
        No_of_owner.append(owner.text)
        rto=driver.find_element_by_xpath("//div[@class='cw-tabs-data']/div/ul/li[10]/div[2]") #item name
        RTO.append(rto.text)
        ins=driver.find_element_by_xpath("//div[@class='cw-tabs-data']/div/ul/li[11]/div[2]") #item name
        Insurance.append(ins.text)
        MRP=driver.find_element_by_xpath("//div[@class='cw-used-cardetails__price-offer']/p") #item name
        Price.append(MRP.text)
```

```python
lst = [3001,3000,40,10,2,176,198,128,12,105,246,224,220,225,160,244,143,222,148,280]
for i in lst:
    url = "https://www.carwale.com/used/cars-for-sale/#sc=-1&so=-1&city="+str(i)+"&pc=+"+str(i)
    process(url)
```
